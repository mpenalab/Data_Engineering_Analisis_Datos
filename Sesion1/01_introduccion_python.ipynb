```python
def saludar(nombre: str) -> str:
    """
    Esta función saluda a un nombre

    Args:
        nombre (str): El nombre de la persona a saludar

    Returns:
        str: Un mensaje de saludo
    """
    print(f"Adios {nombre}")
    
    return f"Hola {nombre}"

var = saludar("Bryan")
print(var)
```
